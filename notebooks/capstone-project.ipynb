## Grid Adventure V1

### Overview

Grid Adventure V1 is, at its core, a grid‑world environment enriched with varied mechanics. The input is a MxN grid (positive integers M, N). Each tile can contain multiple objects (floor, agent, key, door, etc.), each with distinct behavior. The objective is collect any gems if present, before reaching the exit tile.

If you didn’t skip the getting started section at the top of this notebook, you should already be familiar with the game.

### Constraints (Project Scope)

To keep the capstone project tractable, we enforce the following constraints:

- There is exactly 1 agent in the grid, with starting health **at most 5**.
- There is exactly 1 exit tile in the grid.
- The agent moves one tile per step, except under the speed (Boots) power‑up, which doubles movement to two tiles per step.
- Each entity type has a fixed, known cost, damage, or reward (cost = –reward) where applicable:
  - Coins: reward 5.
  - Turn cost: 3 (Every turn incurs a cost of 3 (or reward -3) except for the final turn to win the game).
  - Lava: damage 2.
- Any key can be used to open any door, but each key can only be used once. 
- It is guaranteed that the initial grid will have an **equal number of keys and locked doors**
- It is also guaranteed that any doors would be **locked in the initial state.**
- There are **at most 5** gems in any grid.
- There are **at most 5** key-door pairs in any grid.
- There are **at most 5** pushable boxes in any grid.
- There can be **any number of** coins or lava.
- Power‑up durations are fixed:
  - Boots (Speed): 5 steps.
  - Ghost (Phasing): 5 steps.
  - Shield: 5 uses.
- There are **at most 3** of each powerup type in the grid.
- There will be **no overlapping entities in the initial state** of the grid (i.e each cell contains only `FloorEntity`, or `FloorEntity` with exactly 1 other entity)
- The maximum grid size is in the graded test cases will be **15x15**

### Implementation

````python
"""
Grid Adventure V1 — Optimised Agent
====================================
Key design decisions:

1. MACRO-STATE KEY  — instead of hashing the full immutable State (which
   changes every turn due to score/turn counters), we hash a compact tuple:
     (agent_pos, key_count, hp, gems_left, doors_locked, boxes,
      phase_turns, speed_turns, shield_uses)
   Coins (rewardable) are excluded — they are optional and their inclusion
   multiplies the state space by 2^N for N coins without gating victory.

2. TWO-PHASE SEARCH
   Phase 1 (greedy, 40% of budget): pure heuristic best-first.
   Guaranteed to find *a* solution fast on any solvable level.
   Phase 2 (scored A*, remaining budget): maximises score starting from
   the greedy solution's score as a lower bound for pruning.

3. ADMISSIBLE HEURISTIC
   h = (nearest_gem_dist + farthest_gem_to_exit_dist) * TURN_COST
       + num_gems * TURN_COST (pickup cost)
   Never overestimates → A* finds optimal given enough time.

4. TIME BUDGET  — 30 s for Task 3 (boss), 9 s for Tasks 1/2.
   Caller sets budget; a 0.3 s safety margin is kept.
"""

from grid_adventure.grid import GridState
from grid_adventure.env import ImageObservation
from grid_adventure.step import Action, step
from grid_adventure.grid import to_state
from grid_adventure.movements import MOVEMENTS
from grid_adventure.objectives import OBJECTIVES
from grid_adventure.entities import (
    AgentEntity, FloorEntity, WallEntity, ExitEntity,
    CoinEntity, GemEntity, KeyEntity, LockedDoorEntity,
    LavaEntity, BoxEntity, SpeedPowerUpEntity, ShieldPowerUpEntity,
    PhasingPowerUpEntity, create_agent_entity,
)

import heapq
import time
from collections import deque
import torch
import numpy as np
from PIL import Image

# ---------------------------------------------------------------------------
# CNN model loader (blob injected after training)
# ---------------------------------------------------------------------------
LABEL_MAP = {
    0: FloorEntity,  1: WallEntity,          2: AgentEntity,      3: ExitEntity,
    4: CoinEntity,   5: GemEntity,            6: KeyEntity,        7: LockedDoorEntity,
    8: LavaEntity,   9: BoxEntity,           10: SpeedPowerUpEntity,
    11: ShieldPowerUpEntity, 12: PhasingPowerUpEntity,
}


def get_model(device: str = "cpu", dtype=None):
    import base64, io, zlib
    _blob_b64 = ""

    raw = zlib.decompress(base64.b64decode(_blob_b64))
    m = torch.jit.load(io.BytesIO(raw), map_location=device)
    if dtype is not None:
        dt = getattr(torch, dtype) if isinstance(dtype, str) else dtype
        for p in m.parameters(): p.data = p.data.to(dt)
        for b in m.buffers():    b.data = b.data.to(dt)
    m.eval()
    return m


# ---------------------------------------------------------------------------
# Helpers
# ---------------------------------------------------------------------------

def _max_time(state, store):
    best = 0
    for eid in store:
        tl = state.time_limit.get(eid)
        if tl: best = max(best, tl.amount)
    return best


def _max_usage(state, store):
    best = 0
    for eid in store:
        ul = state.usage_limit.get(eid)
        if ul: best = max(best, ul.amount)
    return best


def _macro_key(state):
    """
    Compact, hashable state descriptor.
    Excludes: turn, score, rewardable (coins), appearance, cost.
    Powerups encoded as integers (remaining turns/uses).
    """
    agent_id  = next(iter(state.agent.keys()))
    agent_pos = state.position.get(agent_id)
    apos = (agent_pos.x, agent_pos.y)

    # Key count: keys with no position are in inventory (picked up)
    key_count = sum(1 for kid in state.key if state.position.get(kid) is None)

    # Current HP
    hp = state.health.get(agent_id)
    cur_hp = hp.current_health if hp else 0

    # Uncollected gem positions
    gems = frozenset(
        (state.position.get(gid).x, state.position.get(gid).y)
        for gid in state.requirable
        if state.position.get(gid) is not None
    )

    # Locked door positions
    doors = frozenset(
        (state.position.get(did).x, state.position.get(did).y)
        for did in state.locked
        if state.position.get(did) is not None
    )

    # Pushable box positions
    boxes = frozenset(
        (state.position.get(bid).x, state.position.get(bid).y)
        for bid in state.pushable
        if state.position.get(bid) is not None
    )

    return (
        apos, key_count, cur_hp,
        gems, doors, boxes,
        _max_time(state, state.phasing),
        _max_time(state, state.speed),
        _max_usage(state, state.immunity),
    )


def _extract_path(node):
    path = []
    while node.parent is not None:
        path.append(node.act)
        node = node.parent
    path.reverse()
    return path


class _Node:
    __slots__ = ("parent", "act", "state")
    def __init__(self, parent, act, state):
        self.parent = parent
        self.act    = act
        self.state  = state
    def __lt__(self, _):
        return False


# ---------------------------------------------------------------------------
# Agent
# ---------------------------------------------------------------------------

class Agent:
    """Hierarchical A* agent for Grid Adventure V1."""

    # Default time budgets (seconds). Boss levels get 30 s.
    _TASK_BUDGET  = 9.0
    _BOSS_BUDGET  = 30.0
    _SAFETY_SLACK = 0.3

    def __init__(self):
        self.route      = []
        self._dist      = {}
        self._exit_pos  = None
        self.model      = get_model()

    # ------------------------------------------------------------------
    # Public step
    # ------------------------------------------------------------------

    def step(self, state) -> Action:
        if self.route:
            return self.route.pop(0)

        if isinstance(state, GridState):
            gs = state
        else:
            gs = self._image_to_gridstate(state)

        # Heuristically pick budget: larger grids / more objects → boss budget
        imut = to_state(gs)
        budget = self._pick_budget(imut)

        self.route = self._plan(imut, budget)
        if not self.route:
            return Action.WAIT
        return self.route.pop(0)

    def _pick_budget(self, state):
        """Use the longer budget when the level looks complex."""
        n_gems  = len(list(state.requirable.keys()))
        n_doors = len(list(state.locked.keys()))
        n_boxes = len(list(state.pushable.keys()))
        area    = state.width * state.height
        if n_gems + n_doors + n_boxes >= 5 or area >= 100:
            return self._BOSS_BUDGET
        return self._TASK_BUDGET

    # ------------------------------------------------------------------
    # Image → GridState
    # ------------------------------------------------------------------

    def _image_to_gridstate(self, img_obs):
        img  = img_obs['image']
        info = img_obs['info']
        gw, gh = info['config']['width'], info['config']['height']
        tw, th = img.shape[1] // gw, img.shape[0] // gh

        tiles = []
        for y in range(gh):
            for x in range(gw):
                tile = img[y*th:(y+1)*th, x*tw:(x+1)*tw]
                t = torch.tensor(
                    np.array(Image.fromarray(tile).resize((64, 64))),
                    dtype=torch.float32
                ).permute(2, 0, 1) / 255.0
                tiles.append(t)

        with torch.no_grad():
            preds = self.model(torch.stack(tiles)).argmax(dim=1)

        gs = GridState(width=gw, height=gh,
                       movement=MOVEMENTS['cardinal'],
                       objective=OBJECTIVES['collect_gems_and_exit'],
                       seed=0)
        idx = 0
        for y in range(gh):
            for x in range(gw):
                etype = LABEL_MAP[preds[idx].item()]
                idx += 1
                if etype is FloorEntity:
                    gs.add((x, y), FloorEntity())
                elif etype is AgentEntity:
                    hp = info['agent']['health']['current_health']
                    gs.add((x, y), create_agent_entity(health=hp))
                else:
                    gs.add((x, y), FloorEntity())
                    gs.add((x, y), etype())
        return gs

    # ------------------------------------------------------------------
    # Planner
    # ------------------------------------------------------------------

    def _plan(self, state, budget: float):
        self._build_dist(state)
        t0 = time.perf_counter()

        # Phase 1: greedy (40% of budget)
        greedy = self._greedy(state, t0, budget * 0.40)

        elapsed = time.perf_counter() - t0
        rem = budget - elapsed - self._SAFETY_SLACK

        if rem <= 0:
            return greedy

        greedy_score = self._simulate_score(greedy, state) if greedy else float('-inf')

        # Phase 2: scored A* (remaining budget)
        return self._scored_astar(state, t0, budget, greedy_score, greedy)

    # ------------------------------------------------------------------
    # Phase 1 — greedy best-first
    # ------------------------------------------------------------------

    def _greedy(self, init_state, t0, budget):
        pq      = []
        visited = set()
        mk = _macro_key(init_state)
        visited.add(mk)
        heapq.heappush(pq, (self._h(init_state), _Node(None, None, init_state)))

        while pq:
            if time.perf_counter() - t0 > budget:
                break
            _, curr = heapq.heappop(pq)
            if curr.state.win:
                return _extract_path(curr)
            for child in self._expand(curr):
                cmk = _macro_key(child.state)
                if cmk in visited:
                    continue
                visited.add(cmk)
                heapq.heappush(pq, (self._h(child.state), child))
        return []

    # ------------------------------------------------------------------
    # Phase 2 — scored A*
    # ------------------------------------------------------------------

    def _scored_astar(self, init_state, t0, total_budget,
                      init_win_score, init_path):
        pq   = []
        best = {}   # macro_key → best score seen at that config

        mk = _macro_key(init_state)
        best[mk] = init_state.score
        heapq.heappush(pq, (self._h(init_state), _Node(None, None, init_state)))

        win_score = init_win_score
        win_node  = None
        deadline  = t0 + total_budget - self._SAFETY_SLACK

        while pq:
            if time.perf_counter() > deadline:
                break

            f, curr = heapq.heappop(pq)

            # Stale?
            mk = _macro_key(curr.state)
            if curr.state.score < best.get(mk, float('-inf')):
                continue

            # Prune: upper bound on achievable score from this node
            # priority f = -score + h  → estimated_upper = score - h + coin_ub
            # Conservatively: upper ≤ -f (ignores future coins, so admissible pruning)
            if win_node is not None and -f <= win_score:
                break

            if curr.state.win:
                if curr.state.score > win_score:
                    win_score = curr.state.score
                    win_node  = curr
                continue

            for child in self._expand(curr):
                cmk    = _macro_key(child.state)
                cscore = child.state.score
                # Only push if strictly better than what we've seen
                if cscore < best.get(cmk, float('-inf')):
                    continue
                best[cmk] = cscore
                heapq.heappush(pq, (-cscore + self._h(child.state), child))

        if win_node is not None:
            return _extract_path(win_node)
        return init_path if init_path else []

    # ------------------------------------------------------------------
    # Heuristic — admissible lower bound on remaining *cost*
    # ------------------------------------------------------------------

    def _h(self, state):
        """
        Returns non-negative estimated cost still to incur.
        Priority = -score + h = -(score collected) + (estimated future spend).
        """
        STEP = 3  # turn cost in reward units

        agent_id  = next(iter(state.agent.keys()))
        agent_pos = state.position.get(agent_id)
        pos       = (agent_pos.x, agent_pos.y)
        is_phase  = _max_time(state, state.phasing) > 0

        gems = []
        for gid in state.requirable:
            gp = state.position.get(gid)
            if gp:
                gems.append((gp.x, gp.y))

        if not gems:
            return self._d(pos, 'exit', is_phase) * STEP

        # Must reach at least nearest gem, then travel to exit from farthest gem.
        # Plus one PICK_UP turn per gem.
        nearest  = min(self._d(pos, g, is_phase)          for g in gems)
        far_exit = max(self._d(g,   'exit', is_phase)     for g in gems)
        return (nearest + far_exit) * STEP + len(gems) * STEP

    # ------------------------------------------------------------------
    # Expand
    # ------------------------------------------------------------------

    def _expand(self, node):
        children = []
        for action in Action:
            if action is Action.WAIT:
                continue
            ns = step(node.state, action)
            if not ns.lose:
                children.append(_Node(node, action, ns))
        return children

    # ------------------------------------------------------------------
    # BFS distance tables
    # ------------------------------------------------------------------

    def _build_dist(self, state):
        w, h = state.width, state.height

        # Walls only (not locked doors, not pushable boxes)
        walls = set()
        for eid in state.blocking:
            if eid not in state.locked and eid not in state.pushable:
                p = state.position.get(eid)
                if p: walls.add((p.x, p.y))

        self._dist = {}

        exit_id  = next(iter(state.exit.keys()))
        ep       = state.position.get(exit_id)
        self._exit_pos = (ep.x, ep.y)
        self._dist['exit'] = self._bfs(self._exit_pos, walls, w, h)

        for gid in state.requirable:
            gp = state.position.get(gid)
            if gp:
                k = (gp.x, gp.y)
                if k not in self._dist:
                    self._dist[k] = self._bfs(k, walls, w, h)

    def _d(self, frm, to, has_phase):
        dmap     = self._dist.get(to, {})
        d_normal = dmap.get(frm, 9999)
        if has_phase:
            dest    = self._exit_pos if to == 'exit' else to
            d_phase = abs(dest[0] - frm[0]) + abs(dest[1] - frm[1])
            return min(d_normal, d_phase)
        return d_normal

    @staticmethod
    def _bfs(start, walls, w, h):
        d = {start: 0}
        q = deque([start])
        while q:
            x, y = q.popleft()
            for nx, ny in ((x-1,y),(x+1,y),(x,y-1),(x,y+1)):
                if 0<=nx<w and 0<=ny<h and (nx,ny) not in walls and (nx,ny) not in d:
                    d[(nx,ny)] = d[(x,y)] + 1
                    q.append((nx,ny))
        return d

    @staticmethod
    def _simulate_score(actions, init_state):
        s = init_state
        for a in actions:
            s = step(s, a)
        return s.score

    # ------------------------------------------------------------------
    # Optional debug stubs
    # ------------------------------------------------------------------

    def parse(self, observation) -> GridState:
        return self._image_to_gridstate(observation)

    def info(self) -> dict:
        return {"name": "Macro-state A* Agent"}
````

### Hardcoding Your Trained ML Model
Because Coursemology disallows arbitrary auxiliary file uploads (e.g. separate `.pth` / `.pkl` weight files), any ML models your `Agent` depends on must ship **inline inside your submission**. To make this easy (and less error‑prone than manually copying numbers), we provide two helper functions in `utils.py`:

- `generate_torch_loader_snippet(model, example_inputs=..., prefer='auto', compression='zlib')`
- `generate_sklearn_loader_snippet(model, compression='zlib')`

Each returns a fully self‑contained Python source string defining a `get_model(...)` function. You copy‑paste that function into your agent code. At runtime the function reconstructs the trained model from a Base64‑encoded (optionally compressed) blob embedded directly in the source.

Below we show the end‑to‑end workflow for PyTorch and scikit‑learn. We also keep (optional) manual hardcoding examples for learning.

Suppose you train a model on the following data:

````python
import numpy as np

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([1, 8, 2, 4])
````

#### PyTorch

Assume you have the following simple network (we'll train it briefly, then auto‑generate a loader snippet):

````python
import torch
from torch import nn

net = nn.Linear(2, 1)
````

Train the network:

````python
def train(net: nn.Module, X: np.ndarray, y: np.ndarray):
    X = torch.tensor(X, dtype=torch.float32)
    y = torch.tensor(y[:, None], dtype=torch.float32)
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(net.parameters(), lr=0.01)
    for epoch in range(100):
        optimizer.zero_grad()
        output = net(X)
        loss = criterion(output, y)
        loss.backward()
        optimizer.step()

train(net, X, y)
````

**Option 1 (Recommended):** Automatically generate a self‑contained loader snippet.

Run the following after training:

````python
# Auto‑generate a loader snippet for the trained PyTorch model
from utils import generate_torch_loader_snippet

example_input = torch.tensor([[0.0, 0.0]], dtype=torch.float32)  # minimal example for tracing if needed
snippet = generate_torch_loader_snippet(net, example_inputs=example_input, prefer="auto", compression="zlib")
print(snippet)

# You would copy the FULL printed function definition (def get_model(...): ...) into your agent source code.
# Then inside Agent.__init__ you could do: self.model = get_model()
````

To hardcode manually:

````python
from collections import OrderedDict
from torch import tensor

# declare the same model
model = nn.Linear(2, 1)

# Hardcode the weights from the previous step (input your values here)
coeffs = OrderedDict([('weight', tensor([[0.1491376609, 1.3386186361]])), ('bias', tensor([0.6359716654]))])

# Load the model with the hardcoded weights
model.load_state_dict(coeffs)

# The model now has the same weights as it was previously trained
X_tensor = torch.tensor(X, dtype=torch.float32)

# Check if the outputs are the same
torch.allclose(model(X_tensor), net(X_tensor))
````
